```python
import pyradon
import numpy as np
import matplotlib.pyplot as plt
from time import perf_counter
```

```python
n_rows = 1200
n_cols = 1200
n_rays = 500
n_slopes = 500
image = np.loadtxt("shepplogan.csv", delimiter = ",").tolist()
```

```python
t_start = perf_counter()
radon_transform = np.array(pyradon.transform(image, n_rows, n_cols, n_rays, n_slopes))
t_stop = perf_counter()
print(f"{(t_stop-t_start)*1e3:.2f} ms")
```

```python
plt.figure(dpi=150, figsize = (6, 3))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap = "gray")
plt.title("Original image")
plt.subplot(1, 2, 2)
plt.imshow(radon_transform.T, cmap = "viridis")
plt.xlabel("Rays")
plt.ylabel("Slopes")
plt.subplots_adjust(wspace = .35)
plt.title("Radon transform")
```
